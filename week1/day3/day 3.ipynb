```sql
SELECT emp_name,
       salary,
       experience,
       performance_rating,
       CASE 
           WHEN experience >= 8 AND performance_rating = 'A' THEN salary * 1.20
           WHEN experience >= 5 AND performance_rating = 'B' THEN salary * 1.15
           WHEN performance_rating = 'C' THEN salary
           ELSE salary * 1.10
       END AS updated_salary
FROM Emp;
```

```sql
SELECT emp_name,
       department,
       salary,
       performance_rating,
       CASE
           -- Finance Department
           WHEN department = 'Finance' AND performance_rating = 'A' THEN salary * 0.20
           WHEN department = 'Finance' AND performance_rating = 'B' THEN salary * 0.15
           WHEN department = 'Finance' AND performance_rating = 'C' THEN salary * 0.05

           -- Engineering Department
           WHEN department = 'Engineering' AND performance_rating = 'A' THEN salary * 0.18
           WHEN department = 'Engineering' AND performance_rating = 'B' THEN salary * 0.12
           WHEN department = 'Engineering' AND performance_rating = 'C' THEN salary * 0.03

           -- Other Departments
           ELSE salary * 0.10
       END AS bonus
FROM Emp;
```

```sql
SELECT emp_name,
       salary,
       performance_rating,
       CASE
           WHEN salary > 80000 AND performance_rating = 'A' 
                THEN 'High Performer'

           WHEN salary BETWEEN 50000 AND 80000 
                AND performance_rating = 'B' 
                THEN 'Mid Performer'

           WHEN salary < 50000 
                OR performance_rating = 'C' 
                THEN 'Low Performer'
       END AS category
FROM Emp;
```

```sql
SELECT emp_name,
       department,
       experience,
       CASE
           -- HR Department
           WHEN department = 'HR' AND experience < 5 
                THEN 'High Risk'
           WHEN department = 'HR' AND experience > 5 
                THEN 'Low Risk'

           -- Engineering & Finance
           WHEN department IN ('Engineering', 'Finance') 
                AND experience > 8 
                THEN 'Low Risk'
           WHEN department IN ('Engineering', 'Finance') 
                AND experience < 8 
                THEN 'Medium Risk'

           -- Other Departments
           ELSE 'Medium Risk'
       END AS risk_level
FROM Emp;
```

```sql
SELECT emp_name,
       salary,
       experience,
       performance_rating,
       CASE
           -- Performance A
           WHEN performance_rating = 'A' THEN
               CASE
                   WHEN salary > 80000 AND experience > 5 THEN salary * 1.25
                   WHEN salary > 80000 THEN salary * 1.20
                   WHEN salary BETWEEN 50000 AND 80000 THEN salary * 1.15
               END

           -- Performance B
           WHEN performance_rating = 'B' THEN
               CASE
                   WHEN experience > 5 THEN salary * 1.12
                   ELSE salary * 1.10
               END

           -- Performance C
           WHEN performance_rating = 'C' THEN salary

       END AS updated_salary
FROM Emp;
```

```sql
SELECT emp_name,
       department,
       salary,
       experience,
       performance_rating,
       CASE
           -- Finance Department
           WHEN department = 'Finance' THEN
               CASE
                   WHEN performance_rating = 'A' AND experience > 10 
                        THEN salary * 0.25
                   ELSE salary * 0.20
               END

           -- HR Department
           WHEN department = 'HR' THEN
               CASE
                   WHEN performance_rating = 'B' OR experience > 5 
                        THEN salary * 0.15
                   ELSE salary * 0.10
               END

           -- Other Departments
           ELSE salary * 0.08
       END AS bonus
FROM Emp;
```

```sql
SELECT emp_name,
       salary,
       experience,
       performance_rating,
       CASE
           -- Salary > 70000
           WHEN salary > 70000 THEN
               CASE
                   WHEN performance_rating = 'A' AND experience > 8 
                        THEN 'Top Performer'
                   WHEN experience < 8 
                        THEN 'Mid Performer'
               END

           -- Salary between 50000 and 70000
           WHEN salary BETWEEN 50000 AND 70000 THEN
               CASE
                   WHEN performance_rating = 'A' 
                        THEN 'Rising Star'
                   ELSE 'Average Performer'
               END

           -- Salary < 50000
           WHEN salary < 50000 THEN 'Low Performer'
       END AS category
FROM Emp;
```

```sql
SELECT emp_name,
       salary,
       experience,
       CASE
           -- Salary > 90000
           WHEN salary > 90000 THEN
               CASE
                   WHEN experience > 10 THEN '35%'
                   ELSE '30%'
               END

           -- Salary between 60000 and 90000
           WHEN salary BETWEEN 60000 AND 90000 THEN
               CASE
                   WHEN experience > 5 THEN '25%'
                   ELSE '20%'
               END

           -- Salary < 60000
           ELSE '15%'
       END AS tax_bracket
FROM Emp;
```

```sql
SELECT emp_name,
       salary,
       experience,
       performance_rating,
       CASE
           -- Performance A
           WHEN performance_rating = 'A' THEN
               CASE
                   WHEN salary > 75000 AND experience > 7 
                        THEN 'Eligible for Senior Role'
                   ELSE 'Eligible for Junior Role'
               END

           -- Performance B
           WHEN performance_rating = 'B' THEN
               CASE
                   WHEN experience > 5 
                        THEN 'Eligible for Consideration'
                   ELSE 'Not Eligible'
               END

           -- Performance C
           WHEN performance_rating = 'C' 
                THEN 'Not Eligible'
       END AS promotion_status
FROM Emp;
```

### ROW_NUMBER() ONLY Questions

```sql
SELECT emp_id,
       emp_name,
       salary,
       ROW_NUMBER() OVER (ORDER BY salary DESC) AS row_num
FROM employee1;
```

```sql
SELECT emp_id,
       emp_name,
       department,
       salary,
       ROW_NUMBER() OVER (PARTITION BY department ORDER BY salary DESC) AS dept_row_num
FROM employee1;
```

```sql
SELECT emp_id,
       emp_name,
       join_date,
       ROW_NUMBER() OVER (ORDER BY join_date DESC) AS row_num
FROM employee1;
```

```sql
SELECT emp_id,
       emp_name,
       department,
       join_date,
       ROW_NUMBER() OVER (PARTITION BY department ORDER BY join_date ASC) AS dept_row_num
FROM employee1;
```

```sql
SELECT emp_id,
       emp_name,
       salary,
       ROW_NUMBER() OVER (ORDER BY salary ASC) AS row_num
FROM employee1;
```

```sql
SELECT emp_id,
       emp_name,
       department,
       ROW_NUMBER() OVER (PARTITION BY department ORDER BY emp_name ASC) AS dept_name_rank
FROM employee1;
```
