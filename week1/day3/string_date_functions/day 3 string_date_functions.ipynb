```sql
CREATE TABLE employee_payments ( 
emp_id INT PRIMARY KEY, 
emp_name VARCHAR(50), 
department VARCHAR(30), 
base_salary DECIMAL(10,2), 
bonus DECIMAL(10,2), 
joining_date DATE 
);
INSERT INTO employee_payments VALUES 
(1,'karthik','Data',75000.75,5000.50,'2019-03-15'), 
(2,'veena','HR',65000.40,4000.25,'2021-06-20'), 
(3,'ravi','Data',85000.90,6000.75,'2016-01-10'), 
(4,'anil','Finance',70000.10,NULL,'2020-09-01'), 
(5,'suresh','HR',60000.55,3000.30,'2022-11-25');
```

```sql
SELECT 
    emp_id,
    INITCAP(emp_name) AS emp_name_proper,

    -- Upper & Lower
    UPPER(emp_name) AS emp_name_upper,
    LOWER(emp_name) AS emp_name_lower,

    department,
    ROUND(base_salary + COALESCE(bonus, 0)) AS total_income,

    -- Extract joining year
    EXTRACT(YEAR FROM joining_date) AS joining_year,

    -- Experience classification
    CASE 
        WHEN EXTRACT(YEAR FROM CURRENT_DATE) - EXTRACT(YEAR FROM joining_date) > 7 
            THEN 'Senior'
        WHEN EXTRACT(YEAR FROM CURRENT_DATE) - EXTRACT(YEAR FROM joining_date) BETWEEN 4 AND 7 
            THEN 'Mid'
        ELSE 'Junior'
    END AS experience_level

FROM employee_payments;
```

```sql
CREATE TABLE orders_delivery ( 
order_id INT, 
customer_name VARCHAR(50), 
order_date DATE, 
delivery_date DATE, 
order_amount DECIMAL(10,2) 
);
INSERT INTO orders_delivery VALUES 
(101,'rajesh','2025-01-01','2025-01-05',12500.75), 
(102,'meena','2025-01-10','2025-01-10',8400.40), 
(103,'arun','2025-01-15','2025-01-20',15600.90), 
(104,'pooja','2025-01-18',NULL,9200.10);
```

```sql
SELECT 
    order_id,

    -- Uppercase customer name
    UPPER(customer_name) AS customer_name_upper,

    order_date,

    -- Replace NULL delivery date with today
    COALESCE(delivery_date, CURRENT_DATE) AS final_delivery_date,

    -- Delivery days calculation
    DATEDIFF(COALESCE(delivery_date, CURRENT_DATE), order_date) AS delivery_days,

    ROUND(order_amount, 1) AS truncated_amount,

    -- Status classification
    CASE 
        WHEN delivery_date IS NULL THEN 'Pending'
        WHEN DATEDIFF(delivery_date, order_date) = 0 THEN 'Same-day'
        WHEN DATEDIFF(delivery_date, order_date) > 3 THEN 'Delayed'
        ELSE 'On-time'
    END AS delivery_status

FROM orders_delivery;
```

```sql
CREATE TABLE customer_spending ( 
cust_id INT, 
cust_name VARCHAR(50), 
city VARCHAR(30), 
purchase_amount DECIMAL(10,2), 
purchase_date DATE 
);
INSERT INTO customer_spending VALUES 
(1,'amit','mumbai',12000.75,'2024-12-01'), 
(2,'neha','delhi',8500.40,'2024-12-15'), 
(3,'rohit','mumbai',15500.90,'2024-11-20'), 
(4,'kavya','chennai',6000.10,'2024-10-05');
```

```sql
SELECT 
    cust_id,

    -- Capitalize first letter
    INITCAP(cust_name) AS cust_name_proper,

    city,

    -- Month name of purchase
    DATE_FORMAT(purchase_date, 'MMMM') AS purchase_month,

    -- Rounded purchase amount
    ROUND(purchase_amount) AS rounded_amount,

    -- Absolute value
    ABS(purchase_amount) AS absolute_amount,

    -- Spending category
    CASE 
        WHEN purchase_amount > 15000 THEN 'High Spender'
        WHEN purchase_amount BETWEEN 8000 AND 15000 THEN 'Medium Spender'
        ELSE 'Low Spender'
    END AS spender_category

FROM customer_spending;
```

```sql
CREATE TABLE subscriptions ( 
user_id INT, 
user_email VARCHAR(100), 
start_date DATE, 
end_date DATE, 
subscription_fee DECIMAL(10,2) 
);
INSERT INTO subscriptions VALUES 
(1,'karthik@gmail.com','2024-01-01','2025-01-01',12000.50), 
(2,'veena@yahoo.com','2024-06-15','2024-12-15',8500.75), 
(3,'ravi@hotmail.com','2023-03-01','2024-03-01',15000.90);
```

```sql
SELECT 
    user_id,

    -- Extract email domain
    SPLIT(user_email, '@')[1] AS email_domain,

    -- Subscription duration in months
    MONTHS_BETWEEN(end_date, start_date) AS duration_months,

    -- Format fee with commas
    FORMAT_NUMBER(subscription_fee, 2) AS formatted_fee,

    -- Remaining days from today
    DATEDIFF(end_date, CURRENT_DATE) AS remaining_days,

    -- Status classification
    CASE 
        WHEN end_date < CURRENT_DATE THEN 'Expired'
        WHEN DATEDIFF(end_date, CURRENT_DATE) <= 30 THEN 'Expiring Soon'
        ELSE 'Active'
    END AS subscription_status

FROM subscriptions;
```

```sql
CREATE TABLE loan_details ( 
loan_id INT, 
customer_name VARCHAR(50), 
loan_amount DECIMAL(12,2), 
interest_rate DECIMAL(5,2), 
loan_start DATE 
);
INSERT INTO loan_details VALUES 
(201,'suresh',500000.75,8.5,'2022-01-10'), 
(202,'mahesh',750000.40,9.2,'2021-05-20'), 
(203,'anita',300000.90,7.8,'2023-07-01'); 
```

```sql
SELECT 
    loan_id,

    -- Uppercase name
    UPPER(customer_name) AS customer_name_upper,

    loan_amount,
    interest_rate,

    -- Monthly interest rate
    (interest_rate / (12 * 100)) AS monthly_interest,

    -- Years since loan start
    FLOOR(MONTHS_BETWEEN(CURRENT_DATE, loan_start) / 12) AS years_since_start,

    -- EMI calculation
    ROUND(
        loan_amount * 
        (interest_rate / (12 * 100)) * 
        POWER(1 + (interest_rate / (12 * 100)), 60) /
        (POWER(1 + (interest_rate / (12 * 100)), 60) - 1)
    ) AS emi,

    -- Risk category
    CASE 
        WHEN interest_rate > 9 THEN 'High Risk'
        WHEN interest_rate BETWEEN 8 AND 9 THEN 'Medium Risk'
        ELSE 'Low Risk'
    END AS risk_category

FROM loan_details;
```

```sql
CREATE TABLE attendance ( 
emp_id INT, 
emp_name VARCHAR(50), 
total_days INT, 
present_days INT, 
record_date DATE 
); 
INSERT INTO attendance VALUES 
(1,'karthik',30,28,'2025-01-31'), 
(2,'veena',30,22,'2025-01-31'), 
(3,'ravi',30,18,'2025-01-31');
```

```sql
SELECT 
    emp_id,

    -- Lowercase employee name
    LOWER(emp_name) AS emp_name_lower,

    total_days,
    present_days,

    -- Attendance percentage 
    ROUND((present_days * 100.0) / total_days) AS attendance_percentage,

    -- Month name
    DATE_FORMAT(record_date, 'MMMM') AS month_name,

    -- Absent days 
    (total_days - present_days) AS absent_days,

    -- Performance category
    CASE 
        WHEN (present_days * 100.0) / total_days >= 90 THEN 'Excellent'
        WHEN (present_days * 100.0) / total_days BETWEEN 75 AND 89 THEN 'Average'
        ELSE 'Poor'
    END AS performance_category

FROM attendance;
```

```sql
CREATE TABLE product_sales ( 
product_id INT, 
product_name VARCHAR(50), 
mrp DECIMAL(10,2), 
selling_price DECIMAL(10,2), 
sale_date DATE 
);
INSERT INTO product_sales VALUES 
(1,'Laptop',75000.75,68000.50,'2025-01-10'), 
(2,'Mobile',35000.40,33000.25,'2025-01-12'), 
(3,'Tablet',25000.90,26000.75,'2025-01-15');
```

```sql
SELECT 
    product_id,

    -- Proper case product name
    INITCAP(product_name) AS product_name_proper,

    mrp,
    selling_price,

    -- Discount amount 
    ABS(mrp - selling_price) AS discount_amount,

    -- Discount percentage
    ROUND(((mrp - selling_price) * 100.0) / mrp, 2) AS discount_percentage,

    -- Day name of sale
    DATE_FORMAT(sale_date, 'EEEE') AS day_name,

    -- Pricing category
    CASE 
        WHEN selling_price < mrp THEN 'Valid Discount'
        WHEN selling_price > mrp THEN 'Overpriced'
        ELSE 'No Discount'
    END AS price_status

FROM product_sales;
```

```sql
CREATE TABLE insurance_policies ( 
policy_id INT, 
holder_name VARCHAR(50), 
premium_amount DECIMAL(10,2), 
policy_start DATE, 
policy_end DATE 
);
INSERT INTO insurance_policies VALUES 
(301,'arjun',12000.50,'2023-01-01','2026-01-01'), 
(302,'megha',8500.75,'2022-06-15','2025-06-15'), 
(303,'vinod',15000.90,'2021-03-01','2024-03-01');
```

```sql
SELECT 
    policy_id,

    -- Uppercase holder name
    UPPER(holder_name) AS holder_name_upper,

    premium_amount,

    -- Rounded premium
    ROUND(premium_amount) AS rounded_premium,

    -- Policy duration in years
    FLOOR(MONTHS_BETWEEN(policy_end, policy_start) / 12) AS policy_duration_years,

    -- Remaining days
    DATEDIFF(policy_end, CURRENT_DATE) AS remaining_days,

    -- Policy category
    CASE 
        WHEN policy_end < CURRENT_DATE THEN 'Expired'
        WHEN MONTHS_BETWEEN(policy_end, policy_start) / 12 >= 3 THEN 'Long Term'
        WHEN MONTHS_BETWEEN(policy_end, policy_start) / 12 BETWEEN 1 AND 3 THEN 'Mid Term'
        ELSE 'Short Term'
    END AS policy_status

FROM insurance_policies;
```

```sql
CREATE TABLE salary_revision ( 
emp_id INT, 
emp_name VARCHAR(50), 
current_salary DECIMAL(10,2), 
rating INT, 
last_hike DATE 
); 
INSERT INTO salary_revision VALUES 
(1,'karthik',75000.75,5,'2023-01-01'), 
(2,'veena',65000.40,4,'2024-01-01'), 
(3,'ravi',85000.90,3,'2022-01-01');
```

```sql
SELECT 
    emp_id,

    -- Lowercase name
    LOWER(emp_name) AS emp_name_lower,

    current_salary,
    rating,
    last_hike,

    -- Years since last hike
    FLOOR(MONTHS_BETWEEN(CURRENT_DATE, last_hike) / 12) AS years_since_hike,

    -- Increment logic based on rating
    CASE 
        WHEN rating = 5 THEN current_salary * 0.20
        WHEN rating = 4 THEN current_salary * 0.15
        WHEN rating = 3 THEN current_salary * 0.10
        ELSE 0
    END AS increment_amount,

    -- New salary
    ROUND(
        current_salary + 
        CASE 
            WHEN rating = 5 THEN current_salary * 0.20
            WHEN rating = 4 THEN current_salary * 0.15
            WHEN rating = 3 THEN current_salary * 0.10
            ELSE 0
        END
    ) AS new_salary,

    -- Performance category
    CASE 
        WHEN rating = 5 THEN 'High Increment'
        WHEN rating = 4 THEN 'Moderate'
        ELSE 'No Increment'
    END AS increment_category

FROM salary_revision;
```

```sql
CREATE TABLE bank_accounts ( 
account_id INT, 
customer_name VARCHAR(50), 
balance DECIMAL(12,2), 
last_transaction DATE, 
branch VARCHAR(30) 
);
INSERT INTO bank_accounts VALUES 
(501,'ramesh',125000.75,'2024-12-20','hyderabad'), 
(502,'sita',8500.40,'2023-06-15','delhi'), 
(503,'manoj',-2500.90,'2025-01-05','mumbai');
```

```sql
SELECT 
    account_id,

    customer_name,

    balance,

    -- Absolute balance
    ABS(balance) AS absolute_balance,

    -- Days since last transaction
    DATEDIFF(CURRENT_DATE, last_transaction) AS days_since_transaction,

    -- Proper case branch
    INITCAP(branch) AS branch_proper,

    -- Sign of balance 
    SIGN(balance) AS balance_sign,

    -- Account status classification
    CASE 
        WHEN balance < 0 THEN 'Overdrawn'
        WHEN DATEDIFF(CURRENT_DATE, last_transaction) > 365 THEN 'Dormant'
        ELSE 'Active'
    END AS account_status

FROM bank_accounts;
```

```sql
CREATE TABLE salary_audit ( 
emp_id INT, 
emp_name VARCHAR(50), 
salary DECIMAL(10,2), 
tax_percent DECIMAL(5,2), 
last_revision DATE 
);
INSERT INTO salary_audit VALUES 
(1,'karthik',75000.75,10.5,'2022-01-15'), 
(2,'veena',65000.40,18.0,'2023-06-01'), 
(3,'ravi',85000.90,25.0,'2020-11-20');
```

```sql
SELECT 
    emp_id,

    -- Normalize name
    LOWER(emp_name) AS emp_name_lower,

    salary,
    tax_percent,
    last_revision,

    -- Net salary after tax
    ROUND(salary - (salary * tax_percent / 100.0)) AS net_salary,

    -- Extract revision year
    EXTRACT(YEAR FROM last_revision) AS revision_year,

    -- Months since revision
    MONTHS_BETWEEN(CURRENT_DATE, last_revision) AS months_since_revision,

    -- Audit flag logic
    CASE 
        WHEN tax_percent > 20 
             AND MONTHS_BETWEEN(CURRENT_DATE, last_revision) > 24 
            THEN 'Tax Shock'

        WHEN tax_percent BETWEEN 15 AND 20 
            THEN 'Review Needed'

        ELSE 'Stable'
    END AS audit_status

FROM salary_audit;
```

```sql
CREATE TABLE bonus_monitor ( 
emp_code INT, 
emp_name VARCHAR(50), 
base_salary DECIMAL(10,2), 
bonus DECIMAL(10,2), 
bonus_date DATE 
);
INSERT INTO bonus_monitor VALUES 
(101,'Anil',70000.10,30000.00,'2025-01-10'), 
(102,'Suresh',60000.55,3000.30,'2024-03-15'), 
(103,'Ravi',85000.90,15000.75,'2023-12-01');
```

```sql
SELECT 
    emp_code,

    -- Proper case name
    INITCAP(emp_name) AS emp_name_proper,

    base_salary,
    bonus,
    bonus_date,

    -- Bonus percentage of salary
    ROUND((bonus * 100.0) / base_salary, 2) AS bonus_percentage,

    -- Day name of bonus
    DATE_FORMAT(bonus_date, 'EEEE') AS bonus_day,

    -- Absolute difference
    ABS(base_salary - bonus) AS salary_bonus_diff,

    -- Weekend check
    CASE 
        WHEN DATE_FORMAT(bonus_date, 'EEEE') IN ('Saturday', 'Sunday') THEN 'Weekend'
        ELSE 'Weekday'
    END AS day_type,

    -- Final classification
    CASE 
        WHEN (bonus * 100.0) / base_salary > 30 
             AND DATE_FORMAT(bonus_date, 'EEEE') IN ('Saturday', 'Sunday')
            THEN 'Suspicious'

        WHEN (bonus * 100.0) / base_salary <= 20
            THEN 'Normal'

        ELSE 'Audit'
    END AS bonus_status

FROM bonus_monitor;
```

```sql
CREATE TABLE employee_experience ( 
emp_id INT, 
emp_name VARCHAR(50), 
joining_date DATE, 
declared_experience INT, 
salary DECIMAL(10,2) 
);
INSERT INTO employee_experience VALUES 
(1,'Veena','2018-07-01',4,65000.40), 
(2,'Ravi','2014-01-10',12,85000.90), 
(3,'Anil','2020-09-01',3,70000.10);
```

```sql
SELECT 
    emp_id,

    -- Uppercase name
    UPPER(emp_name) AS emp_name_upper,

    joining_date,
    declared_experience,
    salary,

    -- Actual experience in years
    FLOOR(MONTHS_BETWEEN(CURRENT_DATE, joining_date) / 12) AS actual_experience,

    -- Difference between declared and actual
    ABS(declared_experience - FLOOR(MONTHS_BETWEEN(CURRENT_DATE, joining_date) / 12)) 
        AS experience_difference,

    -- Floor salary
    FLOOR(salary) AS floored_salary,

    -- Validation logic
    CASE 
        WHEN declared_experience > FLOOR(MONTHS_BETWEEN(CURRENT_DATE, joining_date) / 12)
            THEN 'Overstated'

        WHEN declared_experience < FLOOR(MONTHS_BETWEEN(CURRENT_DATE, joining_date) / 12)
            THEN 'Understated'

        ELSE 'Matched'
    END AS experience_status

FROM employee_experience;
```

```sql
CREATE TABLE salary_digits ( 
emp_id INT, 
emp_name VARCHAR(50), 
salary DECIMAL(10,2), 
credit_date DATE 
);
INSERT INTO salary_digits VALUES 
(1,'Karthik',75000.75,'2025-01-01'), 
(2,'Veena',65000.40,'2025-01-02'), 
(3,'Suresh',60000.55,'2025-01-03');
```

```sql
SELECT 
    emp_id,

    emp_name,
    salary,
    credit_date,

    -- Last two characters of name
    RIGHT(emp_name, 2) AS last_two_chars,

    -- Day of month from credit date
    DAY(credit_date) AS credit_day,

    -- Truncate salary to integer
    CAST(salary AS INT) AS truncated_salary,

    -- MOD of salary 
    MOD(CAST(salary AS INT), 10) AS salary_mod_10,

    -- Pattern matching logic
    CASE 
        WHEN MOD(CAST(salary AS INT), 10) = DAY(credit_date)
            THEN 'Pattern Match'
        ELSE 'No Match'
    END AS match_status

FROM salary_digits;
```

```sql
CREATE TABLE payroll_control ( 
emp_id INT, 
emp_name VARCHAR(50), 
salary DECIMAL(10,2), 
payment_date DATE 
);
INSERT INTO payroll_control VALUES 
(1,'Ravi',85000.90,'2025-01-15'), 
(2,'Anil',70000.10,'2025-01-16'), 
(3,'Veena',65000.40,'2025-01-17');
```

```sql
SELECT 
    emp_id,

    -- Lowercase name
    LOWER(emp_name) AS emp_name_lower,

    salary,
    payment_date,

    -- Weekday name
    DATE_FORMAT(payment_date, 'EEEE') AS weekday,

    -- Rounded salary
    ROUND(salary) AS rounded_salary,

    -- MOD on salary
    MOD(CAST(salary AS INT), 2) AS salary_even_flag,

    -- Weekday parity
    MOD(DAY(payment_date), 2) AS day_parity,

    -- Compliance check
    CASE 
        WHEN MOD(CAST(salary AS INT), 2) = 0 
             AND MOD(DAY(payment_date), 2) = 1
            THEN 'Violation'
        ELSE 'Compliant'
    END AS compliance_status

FROM payroll_control;
```
